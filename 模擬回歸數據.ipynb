```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,50)
y=1.2*x+0.8
plt.scatter(x,y)
plt.plot(x,y,'red')
```

```python
x=np.linspace(0,5,50)
y=1.2*x+0.8+0.3*np.random.randn(50)
plt.scatter(x,y)
```

 開始計算回歸! import scikit-learn

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression() #引進空機器學習線性回歸
```

```python
X=x.reshape(50,1)#機器只會看一個數*50行
```

```python
reg.fit(X,y) #學習X,y之間關係
Y=reg.predict(X)
```

```python
plt.plot(X,Y,'red')#預測的線
plt.scatter(x,y)#觀察點
plt.plot(x,x*1.2+0.8,"green")#真實世界的線
```
